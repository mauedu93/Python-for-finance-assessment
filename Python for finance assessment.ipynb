### Question 3

 Download the price history for AAPL and create a graph over time.

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
aapl = yf.Ticker("AAPL")

hist = aapl.history(period="1d",start=("2020-01-01"))

hist['Close'].plot(color='k',figsize=(16,9))

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }

plt.title("APPLE",fontdict=font)
plt.ylabel("Price",fontdict=font)
plt.xlabel("Years",fontdict=font)
plt.show()

```

#### Question 4

What is the correlation between AAPL and SPY over the past 5 years?

```python
from scipy import stats

spy = yf.Ticker("SPY")

df_aapl = aapl.history(period="1d",start=("2020-01-01"))
df_aapl.rename(columns=df_aapl.iloc[0])

df_spy = spy.history(period="1d",start=("2020-01-01"))
df_spy.rename(columns=df_spy.iloc[0])
                     
c,p = stats.pearsonr(df_aapl['Close'],df_spy['Close'])

print("AAPL vs SPY pearson correlation is: %.2f" %c)
```
